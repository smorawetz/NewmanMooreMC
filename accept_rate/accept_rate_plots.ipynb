```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define plot params

params = {
    "text.usetex": True,
    "font.family": "serif",
    "legend.fontsize": 16,
    "figure.figsize": (10, 3),
    "axes.labelsize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "lines.linewidth": 2,
    "lines.markeredgewidth": 0.8,
    "lines.markersize": 1,
    "lines.marker": "o",
    "patch.edgecolor": "black",
    "pgf.rcfonts": False,
}
plt.rcParams.update(params)
plt.style.use("seaborn-deep")
```

```python
# get data for linear schedule

sizes = [31, 32, 33]
gammas = [0.01, 0.001, 0.0001]

colours = ["red", "green", "blue"]

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(28,5))

for i in range(len(sizes)):
    size = sizes[i]
    ax = axs[i]
    for j in range(len(gammas)):
        gamma = gammas[j]
        accept_rate = np.loadtxt("data/L{0}/gamma{1}_lin_accept_rate.dat".format(size, gamma))
        temp = np.loadtxt("data/L{0}/gamma{1}_lin_temperature.dat".format(size, gamma))
        
        ax.plot(temp, accept_rate, colours[j], label=r"$\gamma = {0}$".format(gamma))
        
    if i == 2:
        plt.legend(frameon=False)
        
    ax.set_title("L = {0}".format(size), fontsize = 24)
    
    fig.text(0.37, 1.05, r"Acceptance rate for a linear schedule, $T = 1 - \gamma t$", fontsize=32)
    
    fig.savefig("lin_comparison.pdf", bbox_inches="tight")
```

```python
# get data for exponential schedule

sizes = [31, 32, 33]
gammas = [0.1, 0.01, 0.001]
stoptimes = [50, 500, 5000]

colours = ["red", "green", "blue"]

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(28,5))

for i in range(len(sizes)):
    size = sizes[i]
    ax = axs[i]
    for j in range(len(gammas)):
        gamma = gammas[j]
        stoptime = stoptimes[j]
        accept_rate = np.loadtxt("data/L{0}/gamma{1}_stoptime{2}_expon_accept_rate.dat".format(size, gamma, stoptime))
        temp = np.loadtxt("data/L{0}/gamma{1}_stoptime{2}_temperature.dat".format(size, gamma, stoptime))
        
        ax.plot(temp, accept_rate, colours[j], label=r"$\gamma = {0}$".format(gamma))
        
    if i == 2:
        plt.legend(frameon=False)
        
    ax.set_title("L = {0}".format(size), fontsize = 24)
    
    fig.text(0.34, 1.05, r"Acceptance rate for an exponential schedule, $T = e^{-\gamma t}$", fontsize=32)

        
    fig.savefig("expon_comparison.pdf", bbox_inches="tight")
```
